```python
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# Read data
df = pd.read_csv('/Users/nezihyalabik/Desktop/real_data.csv')
```

```python
df.describe()
```

```python
# Data visualization
plt.figure(figsize=(15, 8))

plt.scatter(df['avg_daily'], df['avg_los'], s=50)
plt.xlabel("ADR")
plt.ylabel("ALOS")
plt.show()
```

```python
plt.boxplot(df['avg_daily'])
```

```python
# Removing outliers
df2 = df[(df.avg_daily < 600)& (df.avg_los < 10)]

plt.scatter(df2['avg_daily'], df2['avg_los'], s=50)
plt.xlabel("ADR")
plt.ylabel("ALOS")
plt.show()
```

```python
# Scale values
scaler = MinMaxScaler()
df2[['avg_daily', 'avg_los']] = scaler.fit_transform(df2[['avg_daily', 'avg_los']])
```

```python
# Elbow technique
sse = {}
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k).fit(df2)
    sse[k] = kmeans.inertia_ # Inertia: This is SSE defined above
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel("Number of clusters")
plt.ylabel("SSE")
plt.subplot(122)
plt.scatter(df2[:, 0], df2[:, 1]);
```

```python
# Number of cluster selection
plt.figure(figsize=(15, 8))

for i in range(2, 11):
    kmeans = KMeans(n_clusters=i)

    kmeans.fit(df2[['avg_daily', 'avg_los']])
    y_kmeans = kmeans.predict(df2[['avg_daily', 'avg_los']])
    centers = kmeans.cluster_centers_
    
    plt.subplot(3, 3, i - 1)
    plt.title('Cluster: {}'.format(i))
    plt.scatter(df2['avg_daily'], df2['avg_los'], c=y_kmeans, s=50, cmap='viridis')
    plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.tight_layout()
plt.xlabel("ADR")
plt.ylabel("ALOS")
plt.show()
```

```python
# Apply kmeans
kmeans = KMeans(n_clusters=20)

kmeans.fit(df2[['avg_daily', 'avg_los']])
y_kmeans = kmeans.predict(df2[['avg_daily', 'avg_los']])
centers = kmeans.cluster_centers_
```

```python
# Kmeans visualization
plt.figure(figsize=(15, 8))

plt.scatter(df2['avg_daily'], df2['avg_los'], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.xlabel("ADR")
plt.ylabel("ALOS")
plt.show()
```
